```python
import tkinter as tk
from tkinter import messagebox

rice = {"Jollof Rice": 350, "Coconut Fried Rice" : 350, "\nJollof Spaghetti": 350}
sides = {"Savory Beans": 350, "Roasted Sweet Potatoes": 300, "Fried Plantains": 150, "Mixed Vegetable Salad" : 150, "Boiled Yam": 150}
swallow = {"Eba": 100, "Poundo Yam": 100, "Semo" : 100, "Atama Soup": 450, "Egusi Soup": 480}
protein = {"Sweet Chilli Chicken": 1100, "Grilled Chicken Wings": 400, "Fried Beef": 400, "Fried Fish": 500, "Boiled Egg": 200, "Sauteed Sausages": 200}
beverages = {"Water": 200, "Glass Drink (35cl)": 150, "PET Drink (35cl)": 300, "PET Drink (50cl)": 350, "Glass/Canned Malt": 500, "Fresh-Yo": 600,
             "Pineapple Juice": 350, "Mango Juice": 350, "Zobo Drink": 350}

menu = {"Jollof Rice": 350, 
            "Coconut Fried Rice" : 350, 
            "Jollof Spaghetti": 350, 
            "Savory Beans": 350, 
            "Roasted Sweet Potatoes": 300,
            "Fried Plantains": 150, 
            "Mixed Vegetable Salad" : 150, 
            "Boiled Yam": 150, 
            "Eba": 100, 
            "Poundo Yam": 100, 
            "Semo" : 100,
            "Atama Soup": 450, 
            "Egusi Soup": 480, 
            "Sweet Chilli Chicken": 1100, 
            "Grilled Chicken Wings": 400, 
            "Fried Beef": 400, 
            "Fried Fish": 500, 
            "Boiled Egg": 200, 
            "Sauteed Sausages": 200, 
            "Water": 200, 
            "Glass Drink (35cl)": 150, 
            "PET Drink (35cl)": 300, 
            "PET Drink (50cl)": 350, 
            "Glass/Canned Malt": 500, 
            "Fresh-Yo": 600,
            "Pineapple Juice": 350, 
            "Mango Juice": 350, 
            "Zobo Drink": 350}



def Cafeteria(name,order,total):
    window = tk.Toplevel(root)
    window.title("Menu")
    window.geometry("1000x1000")
    
         
    label_1 = tk.Label(window, text=f"{name}")
    label_1.pack()

    label_2 = tk.Label(window, text=f"Your order is: {order}")
    label_2.pack()

    label_3 = tk.Label(window, text=f"Total: ₦{total}")
    label_3.pack()

    label_4 = tk.Label(window, text="Thank you!")
    label_4.pack()
    
def enter():
    name = name_entry.get()
    order = order_entry.get()
    quantity = int(quantity_entry.get())

    if order in menu:
        price = menu[order]
        sum = price * quantity
    

        if sum < 2500:
            total = sum - (sum * 0.1) 
            Cafeteria(name,order,total)
        elif sum > 2500 and sum < 5000:
            total = sum - (sum * 0.15)
            Cafeteria(name,order,total)
        elif sum > 5000:
            total = sum - (sum * 0.25)
            Cafeteria(name,order,total)
        elif sum < 1000:
            total = sum
            Cafeteria(name,order,total)
        else:
            total = sum
            Cafeteria(name,order,total)
    else:
        messagebox.showerror("Error", "The selected order is not available in the menu.")


root = tk.Tk()
root.title("Menu")
root.geometry("500x200")

rice_label = tk.Label(root, text= f"RICE/SPAGHETTI \n{rice}")
rice_label.pack()

sides_label = tk.Label(root, text= f"SIDES \n{sides}")
sides_label.pack()

swallow_label = tk.Label(root, text= f"SWALLOW \n{swallow}")
swallow_label.pack()

protein_label = tk.Label(root, text= f"PROTEIN \n{protein}")
protein_label.pack()

beverages_label = tk.Label(root, text= f"BEVERAGES \n{beverages}")
beverages_label.pack()

root = tk.Tk()
root.title("Menu")
root.geometry("500x200")

name_label = tk.Label(root, text= "Name")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

order_label = tk.Label(root, text= "Order")
order_label.pack()
order_entry = tk.Entry(root)
order_entry.pack()

quantity_label = tk.Label(root, text= "Quantity")
quantity_label.pack()
quantity_entry = tk.Entry(root)
quantity_entry.pack()

enter_button = tk.Button(root, text= "Enter", command=enter)
enter_button.pack()
enter_button.config(fg= "green")
root.mainloop()
```
